```python
# Write the DataFrame to an Excel file
output_file = 'app_info.xlsx'
with pd.ExcelWriter(output_file) as writer:
    df.to_excel(writer, sheet_name='App Info', index=False)

print(f"Excel file has been created: {output_file}")
    
```
